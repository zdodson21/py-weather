# API Call Limit Calculator

Calculate how many API calls will be made in a 24 hour time period

```python
# Number of API calls per increment (do not change unless you know what you are doing)
num_of_api_calls_per_increment = 3 # Default 3

# How many minutes between API call(s)?
time_increment_minutes = 5

# Using API for Geocode data?
api_geocode = True

if (not api_geocode):
    num_of_api_calls_per_increment = 2

#######################################################################################

calls_per_hour = 60 / time_increment_minutes

print('Number of free API calls per day: 1000')
print(f'Number of API calls per hour: {round(calls_per_hour * num_of_api_calls_per_increment )}')
print(f'Number of API calls per day: {round(calls_per_hour * num_of_api_calls_per_increment * 24)}')
```
